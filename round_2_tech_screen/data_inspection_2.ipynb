Which time period has the highest average points_earned?

```python
avg_pts_month = df_points.groupby(df_points['CREATED_DATE'].dt.month).mean('POINTS_EARNED')
avg_pts_month.sort_values('POINTS_EARNED', ascending=False)
```

```python
avg_pts_date = df_points.groupby(df_points['CREATED_DATE'].dt.date).mean('POINTS_EARNED')
avg_pts_date.sort_values('POINTS_EARNED', ascending=False)
```

```python
avg_pts_dow = df_points.groupby(df_points['CREATED_DATE'].dt.dayofweek).mean('POINTS_EARNED')
avg_pts_dow.sort_values('POINTS_EARNED', ascending=False)
```

Which transaction type generates the highest average points_earned?

```python
avg_pts_type = df_points.groupby(df_points['TRANSACTION_TYPE']).mean('POINTS_EARNED')
avg_pts_type.sort_values('POINTS_EARNED', ascending=False)
```

Which transaction type has the highest points_earned in total?

```python
sum_pts_type = df_points.groupby(df_points['TRANSACTION_TYPE']).sum('POINTS_EARNED')
sum_pts_type.sort_values('POINTS_EARNED', ascending=False)
```
